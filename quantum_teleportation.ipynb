```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
```

```python
%matplotlib inline
```

```python
circuit.draw(output='mpl')
```

```python
circuit.x(0)
```

```python
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(1)
```

```python
circuit.cx(1,2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
```

```python
circuit.h(0)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
```

```python
circuit.measure([0,1],[0,1])
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
```

```python
circuit.cx(1,2)
```

```python
circuit.cz(0,2)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.measure(0,2)
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(circuit, backend = simulator, shots=1024).result()
```

```python
counts = result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(counts)
```
